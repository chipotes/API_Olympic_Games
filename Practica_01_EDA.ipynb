```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
games_S = pd.read_csv('Data/Athletes_summer_games.csv')
games_W = pd.read_csv('Data/Athletes_winter_games.csv')
regions = pd.read_csv('Data/regions.csv')
df = pd.concat([games_S, games_W])
del games_S, games_W
```

```python
df.info()
```

```python
int(df['Age'].mean())
```

```python
diccionario_nulos = {'Age':int(df['Age'].isnull().sum())}
diccionario_nulos['Sex']=int(df['Sex'].isnull().sum())
diccionario_nulos['Team']=int(df['Team'].isnull().sum())
diccionario_nulos['NOC']=int(df['NOC'].isnull().sum())
diccionario_nulos['Games']=int(df['Games'].isnull().sum())
diccionario_nulos['Year']=int(df['Year'].isnull().sum())
diccionario_nulos['Season']=int(df['Season'].isnull().sum())
diccionario_nulos['City']=int(df['City'].isnull().sum())
diccionario_nulos['Sport']=int(df['Sport'].isnull().sum())
diccionario_nulos['Event']=int(df['Event'].isnull().sum())
diccionario_nulos['Medal']=int(df['Medal'].isnull().sum())
diccionario_nulos

```

```python
diccionario_nulos['City']=int(df['City'].isnull().sum())
```

```python
diccionario_nulos
```

```python
df=df.drop(columns="Unnamed: 0")
```

```python
df["Age"]=df["Age"].fillna(df["Age"].mean())
df["Age"]=df["Age"].astype(int)
```

```python
df.info()
```

```python
variables=list(diccionario_nulos.keys())
nulos=list(diccionario_nulos.values())
plt.bar(variables,nulos)
plt.xlabel("Variables")
plt.ylabel("Cantidad de nulos")
plt.title("Total de nulos por variable")
plt.xticks(rotation=90)
plt.show()
```
